## String-Enclosed Character

```python
breakfast='bacon eggs' # single quotes
message="I don't like spiders" # double quotes
message='I don\'t like spiders' # Use \' to escape the single quote
```

```python
message
```

```python
print('"Isn\'t," she said.')
s = 'First line.\nSecond line.' # \n means newline
print(s) # with print(), \n produces a new line
```

```python
print('C:\some\name') # here \n means newline!
```

```python
print(r'C:\some\name') # note the r before the quote
# Strings will be printed as raw strings if r is added
```

```python
message="""Dear Diary,
I have stomachache today.
It sucks."""

print (message)
```

## String Methods

```python
'ready'.capitalize()
```

```python
'WOW'.casefold()
```

```python
'Python'.center(20)
```

```python
txt='kiwi bird does not eat kiwi fruit'
txt.count('kiwi')
```

```python
txt='Joey is joey'
txt.encode() # Returns a utf-8 encoded version of the string
```

```python
txt.endswith('y')
```

```python
txt = "H\te\tl\tl\to"

x =  txt.expandtabs(2)

print(x)
```

```python
txt = "H\te\tl\tl\to"

x =  txt.expandtabs(3)

print(x)
```

```python
txt='I am eating because I am hungry'
txt.find('am')
```

```python
txt = "For only {price:.2f} dollars!"
print(txt.format(price = 49))
```
