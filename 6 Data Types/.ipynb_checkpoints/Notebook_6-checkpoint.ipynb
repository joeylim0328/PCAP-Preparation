A list is a collection which is ordered and changeable.

```python
flowers=['daisy','rose','tulip']
print(flowers)
```

```python
flowers=['daisy','rose','tulip']
print(flowers[1])
```

```python
flowers[1]='hibiscus'
```

```python
print(flowers)
```

```python
# Loop through a list
flowers=['daisy','rose','tulip']
for flower in flowers:
    print(flower)
```

```python
# Check if item exits
flowers=['daisy','rose','tulip']
if "rose" in flowers:
    print('Exist')
    
else:
    print("Does not exist")
```

```python
# List length
flowers=['daisy','rose','tulip']
print(len(flowers))
```

```python
flowers.append('sunflower')
flowers
```

```python
flowers.insert(1, "magnolia")
flowers
```

```python
# Remove item
flowers.remove('magnolia')
flowers
```

```python
flowers.pop()
flowers
```

```python
del flowers[0]
flowers
```

## Tuple
A tuple is a collection which is ordered and unchangeable.

```python
fruits = ("apple", "banana", "cherry")
print(fruits)
```

```python
# Access items
fruits[2]
```

```python
# Loop through a tuple
fruits = ("apple", "banana", "cherry")
for fruit in fruits:
    print(fruit)
```

```python
# Check if item exists
fruits = ("apple", "banana", "cherry")
if "cherry" in fruits:
    print('Exist')
    
else:
    print("Does not exist")
```

```python
# Tuple length
print(len(fruits))
```

```python
# tuple() constructor
fruits = tuple(("apple", "banana", "cherry"))
fruits
```

## Set
A set is a collection which is unordered and unindexed.

```python
fruits = {"apple", "banana", "cherry"}
print(fruits)
```

```python
# Loop through a set
for fruit in fruits:
    print(fruit)
```

```python
# Check if item exists

if "apple"in fruits:
    print('Exist')
else:
    print('Does not exist')
```

```python
# Length
print(len(fruits))
```

```python
fruits.add("orange")
```

```python
# Add multiple items
fruits.update(["orange","mango","grapes"])
fruits
```

```python
fruits.remove("banana")
fruits
```

```python
fruits.pop()
fruits
```

```python
fruits.discard("grapes")
fruits
```

```python
# Clear
fruits.clear()
fruits
```

```python
fruits={'apple','banana','cherry'}
del fruits # Delete the set completely
```

## Dictionaries
A dictionary is a collection which is unordered, changeable and indexed.

```python
mycar={"brand":"Proton",
      "model":"Saga",
      "year":"2012"}
print(mycar)
```

```python
# Change item
mycar["year"]="2013"
mycar
```

```python
# Accessing items
print(mycar["model"])
print(mycar.get("model"))
```

```python
# Loop to get keys
for key in mycar:
    print(key)
```

```python
# Loop to get values
for key in mycar:
    print(mycar[key])
```

```python
# Loop to return values and items
for value in mycar.values():
    print(value)
for key, value in mycar.items():
    print(key, value)
```

```python
# Check if item exists

if "model" in mycar:
    print('Exists')
    
else:
    print("Does not exist")
```

```python
# List length
print(len(mycar))
```

```python
# Add item
mycar["color"]="red"
mycar
```

```python
# Remove item
mycar.pop("model")
mycar
```

```python
# Pop item removes the lastly inserted value
mycar.popitem()
mycar
```

```python
mycar = { "brand":"Proton", "model":"Saga", "year":"2012","color":"red" }
del mycar["color"]
mycar
```

```python
mycar.clear()
```
